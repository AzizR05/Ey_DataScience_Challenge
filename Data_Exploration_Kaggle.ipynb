```python
#Buliding the Generator and Neural Network
import numpy as np
import ast

def csv_generator(your_list,bs):
    i = 0
    while True:
        x_features = []
        y_features = []
        if i >= len(your_list):
            i = 0
        if len(your_list[i]) != 0 and i < len(your_list):
            x1 = ast.literal_eval(your_list[i][0])
            x1 = [n for n in x1]
            x_features.append(np.array(x1))
            
            y1 = ast.literal_eval(your_list[i][1])
            y1 = [n for n in y1]
            y_features.append(np.array(y1))
            i+=1
        else:
            i+=1
            continue
        yield(np.array(x_features),np.array(y_features))
```

```python
# root mean squared error (rmse) for regression (only for Keras tensors)
def rmse(y_true, y_pred):
    from keras import backend
    return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis=-1))

# mean squared error (mse) for regression  (only for Keras tensors)
def mse(y_true, y_pred):
    from keras import backend
    return backend.mean(backend.square(y_pred - y_true), axis=-1)

# coefficient of determination (R^2) for regression  (only for Keras tensors)
def r_square(y_true, y_pred):
    from keras import backend as K
    SS_res =  K.sum(K.square(y_true - y_pred)) 
    SS_tot = K.sum(K.square(y_true - K.mean(y_true))) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,TimeDistributed
import csv

model = Sequential()
model.add(LSTM(64,activation='relu',return_sequences=True,input_shape=(None,4)))
model.add(LSTM(32,activation='relu'))
model.add(Dense(2,activation='linear'))

with open('normalized_train_data_nn.csv','r') as f:
    reader = csv.reader(f)
    your_list = list(reader)
f.close()

trainGen = csv_generator(your_list[:int(len(your_list)*0.8)],1)
valGen = csv_generator(your_list[int(len(your_list)*0.8):],1)
model.load_weights("model6.h5")
model.compile(loss ='mean_squared_logarithmic_error',optimizer='adam',metrics=['accuracy', rmse,mse,r_square])
model.fit_generator(trainGen,steps_per_epoch=(len(your_list[:int(len(your_list)*0.8)])-1),validation_data=valGen,validation_steps=(len(your_list[int(len(your_list)*0.8):])-1),epochs = 1,verbose=1)
```
